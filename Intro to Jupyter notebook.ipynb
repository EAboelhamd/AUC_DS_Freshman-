### 2. Using (pip) command: 

This way is recommended if you want to install only Python without any other programming lanuages

Just follow the following steps to install python using "pip":

#### 1. Upgrade pip (using the one of the below commands) 

On Windows

    python -m pip install -U pip setuptools
    
On OS X or Linux

    pip install -U pip setuptools
    
    
#### 2. Run one of the below commands to install any desired version of python (python 3 is recommended) 

Python2

    pip install jupyter

Python 3

    pip3 install jupyter
